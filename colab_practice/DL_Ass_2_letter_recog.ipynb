```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix as cm1
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
```

```python
dataset = pd.read_csv("sample_data/letter-recognition.data",sep=",")
```

```python
dataset
```

```python
X = dataset.iloc[:,1:17]
Y = dataset.select_dtypes(include = [object])
```

```python
X
```

```python
Y
```

```python
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X,Y,test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_validation = scaler.transform(X_validation)
Y_train = Y_train.values.ravel()
Y_validation = Y_validation.values.ravel()
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(128,64),max_iter=300,activation='relu',solver='adam',early_stopping=True,verbose=1,random_state=42)

mlp.fit(X_train,Y_train)
```

```python
Y_pred = mlp.predict(X_validation)
print("accuracy:",classification_report(Y_validation, Y_pred))
```

```python
cm = confusion_matrix(Y_validation, Y_pred, labels=list("ABCDEFGHIJKLMNOPQRSTUVWXYZ"))
cm
```

```python
print("Accuracy: ", accuracy_score(Y_validation, Y_pred))
```
